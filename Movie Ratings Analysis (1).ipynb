```python

import pandas as pd
```

```python
df = pd.read_csv('Rotten Tomatoes Movies.csv')
df.head()
```

```python
df['audience_count']
```

```python
df.isna().sum()
```

```python
import numpy as np
```

```python
df =df.dropna(subset=('movie_info','critics_consensus'))
```

```python
df.rating.value_counts()
```

```python
df['genre'] = np.where(df['genre'].isna(), "Unknown Genre" ,df['genre'])
```

```python
df.rating = np.where(df.rating == 'PG-13)', 'PG-13', df.rating)
```

```python
df.rating = np.where(df.rating == 'R)', 'R', df.rating)
```

```python
df
```

```python
df.directors = np.where(df.directors.isna(), 'Unknown', df.directors)
df.writers = np.where(df.writers.isna(), 'Unknown', df.writers)
```

```python
df.cast = np.where(df.cast.isna(), 'Unknown',df.cast)
```

```python
df.info()
```

```python
df.in_theaters_date
```

```python
df['in_theaters_date'] = pd.to_datetime(df['in_theaters_date']) 
df['on_streaming_date'] = pd.to_datetime(df['on_streaming_date']) 
```

```python
df =df.dropna(subset=("in_theaters_date","on_streaming_date"))
```

```python
df.runtime_in_minutes = np.where(df.runtime_in_minutes.isna(),df.runtime_in_minutes.mean(), df.runtime_in_minutes)
```

```python
df.studio_name = np.where(df.studio_name.isna(),"Unknown", df.studio_name)
```

```python
df.audience_rating = np.where(df.audience_rating.isna(),df.audience_rating.mean(), df.audience_rating )
```

```python
df.audience_count = df.audience_count.fillna(df.audience_count.mean())
```

```python
df.tomatometer_status.value_counts().plot.barh(title="Total Tomato-meter Status Count").set_ylabel('');
```
